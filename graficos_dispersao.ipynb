```python
# Gráficos de Dispersão
# Ideal para comparar duas variaveis continuas, duas variáveis numéricas
```

```python
# importando as bibliotecas
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Gráfico de dispersão considerando o volume e a circunferência
plt.scatter(base.Girth, base.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
# Gráfico de linha considerando o volume e circunferencia
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
# Utilizando o seaborn
# Gráfico de dispersão com afastamento dos dados(jitter)
# fit_reg, linha de tendência
sb.regplot(base.Girth, base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```
