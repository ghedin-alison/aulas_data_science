```python
# Importar as bibliotecas
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
sb.scatterplot(base.conc, base.uptake, hue = base.Type)
```

```python
# Seleção de registros especificos da base de dados(Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
m.head()
```

```python
# Subgrafico(1 linha e duas colunas mostrando graficos sobre cada região)
plt.figure()
plt.subplot(1,2,1)
sb.scatterplot(q.conc, q.uptake, color = 'red').set_title('Quebec')
plt.subplot(1,2,2)
sb.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
# Seleção de registros especificos da base de dados(refrigrado e não-refrigerado)
ct = base.loc[base['Treatment'] == 'chilled']
nct = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
sb.scatterplot(ct.conc, ct.uptake, color = 'red').set_title('Chilled Treatment')
plt.subplot(1,2,2)
sb.scatterplot(nct.conc, nct.uptake).set_title('Non Chilled Treatment')
plt.tight_layout()
```

```python
# Carregamento da base de dados de cancer de esofago
base_esofago = pd.read_csv('esoph.csv')
base_esofago.head()
```

```python
# Grafico entre os atributos alcgp e ncontrols
sb.catplot(x = 'alcgp', y = 'ncontrols', data = base_esofago, jitter = False)
```

```python
# Grafico entre os atributos alcgp, ncontrols com agrupamento
sb.catplot(x = 'alcgp', y = 'ncontrols', data = base_esofago, col = 'tobgp')
```
