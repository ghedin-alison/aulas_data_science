```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.shape
```

```python
base.head()
```

```python
base.tail()
```

```python
# Criação d histograma, considerando somente o segundo atributo da base de dados e com duas divisões(bins)
# A variável 'h' armazena as faixas de valores de height
h = np.histogram(base.iloc[:,1], bins = 6) # visualização de como ficará o q o plt vai acabar fazendo
h
```

```python
# Visualização do histograma com 6 divisões
plt.hist(base.iloc[:,1], bins = 6) # plotar histograma na coluna 1 com 6 eixos. Cada eixo é um resumo do intervalo de dados em que está contido.
plt.title('Árvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```
