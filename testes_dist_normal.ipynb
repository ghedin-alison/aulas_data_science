```python
# Testes de distribuição normal
# Com frequencia é necessário verificar se os dados estão normalmente distribuidos
```

```python
# Importação de bibliotecas
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma variavel com dados em uma distribuição normal com a função rvs(100 elementos)
dados = norm.rvs(size = 1000)
```

```python
# histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Geração de gráfico para verificar se a distribuição é normal(qqplot)
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# Execução do teste de Shapiro-Wilk
# Segundo argumento é o valor de p, se (pvalue) > 0.05, não há como rejeitar a hipotese nula.
# hipotese nula(é a hipotese de que os dados estão normalmente distribuidos)
stats.shapiro(dados)
```

```python
# Dados não normais
```

```python
dados2 = skewnorm.rvs(4, size = 1000) # skewnorm cria dados não normais
```

```python
# Geração de gráfico para verificar se a distribuição é normal(qqplot)
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
# Teste de Shapiro-Wilk
stats.shapiro(dados2)
# pvalue < 0.05(1,01 e-15)
```

```python
# histograma
plt.hist(dados2, bins = 20)
plt.title('Dados 2')
```
