```python
# Redes Neurais artificiais com keras
```

```python
# Import das bibliotecas
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
# keras é uma abstração do tensorflow
```

```python
# Carregamento da base de dados e criação dos previsores (variáveis independentes - x) e (variáveis dependentes - y)
base = datasets.load_iris()
previsores = base.data
classe = base.target
classe
```

```python
# Transformação da classe para o formato "dummy", pois temos uma rede neural com 3 neuronios na camada de saida
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
# Divisão da base de dados entre treinamento e teste(70% treino, 30% teste)
x_treina, x_teste, y_treina, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 0)
```

```python
# Criação da estrutura da rede neural com a classe Sequential(sequencia de camadas)
modelo = Sequential()
# Primeira camada oculta, 5 neuronios, 4 neuronios de entrada
modelo.add(Dense(units = 5, input_dim = 4))
# Segunda camada oculta
modelo.add(Dense(units = 4))
# Terceira camada oculta
# Função softmax porque temos um problema de classificação com mais de duas classes
# (é gerada uma probabilidade em cada neuronio)
modelo.add(Dense(units = 3, activation = 'softmax'))
```

```python
# Visualização da estrutura da rede neural
modelo.summary()
```

```python
# Configuração dos parâmetros da rede neural(adam = algoritmo para atualizar os pesos e loss = cálculo do erro)
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', 
               metrics = ['accuracy'])
# Treinamento, dividindo a base de treinamento em uma porção para validação(validation data)
modelo.fit(x_treina, y_treina, epochs = 1000,
           validation_data = (x_teste, y_teste))
```

```python
# Previsões e mudar a variável para True ou False de acordo com o threshold 0.5
previsoes = modelo.predict(x_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Como é um problema com 3 saídas, precisamos buscar a posição que possui o maior valor(são retornados 3 valores)
y_teste_matriz = [np.argmax(t) for t in y_teste]
y_previsao_matriz = [np.argmax(t) for t in previsoes]
```

```python
# Geração da matriz de confusão
confusao = confusion_matrix(y_teste_matriz, y_previsao_matriz)
confusao
```
