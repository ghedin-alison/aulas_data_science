```python
# Importação das bibliotecas
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
sb.distplot(base.Volume, bins = 10, axlabel = 'Volume').set_title('Arvores')
```

```python
# Carregamento da outra base de dados
base2 = pd.read_csv('chicken.csv')
base2.head()
```

```python
agrupado_base2 = base2.groupby(['feed'])['weight'].sum()
agrupado_base2
```
