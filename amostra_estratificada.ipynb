```python
# Importar biblioteca pandas e a funcao train_test_split do sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Carregamento da base de dados e contagem de registros por classe
base = pd.read_csv('iris.csv')
# contar a ocorrencia da coluna classe
base['class'].value_counts()
```

```python
# selecionar uma amostragem estratifica com representação proporcional, mas com aleatoridade
# A funcao train_test_split retorna 4 valores, mas só nos interessam dois (X e Y), os outros dois, utilizamos underline
# base.iloc[:, 0:4] : buscamos somente os atributos previsores, os dados sobre petala e sepala da planta - previsores
# base.iloc[:, 4] : buscamos somente a classe, que é a espécie da planta(versicolor, virginica, setosa) - variavel dependente
# test_size : selecionamos 50% da base de dados, que serão copiados para as variaveis X e Y. Essa função retorna quatro valores,
# porém, vamos usar somente os 50% da base de dados e por isso colocamos "_" para os outros valores
#  stratify : para retornar a amostra baseada na classe


X, _, Y, _ = train_test_split(base.iloc[:, 0:4], base.iloc[:, 4], test_size = 0.5, stratify = base.iloc[:,4])
Y.value_counts()
```
