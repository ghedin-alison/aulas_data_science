```python
# Import de bibliotecas
import matplotlib.pyplot as plt
import nltk # biblioteca de processo de linguagem natural
nltk.download('stopwords') # palavras sem valor semantico(no portugues: para, do, a)
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
# Criação de um corpus. Lendo textos do diretorio
corpus = PlaintextCorpusReader('Arquivos', '.*', encoding = 'ISO-8859-1')
print(corpus)
```

```python
# Leitura dos arquivos do diretorio, percorrer os registros e mostrar o nome dos primeiros 100 arquivos
arquivos = corpus.fileids()
# primeiro aquivo
arquivos[10]
```

```python
# imprime todos os nomes
for a in arquivos:
    print(a)
```

```python
# Acesso ao texto do primeiro arquivo
texto = corpus.raw(arquivos[0])
texto
```

```python
# Acesso de todas as palavras de todos os arquivos
todo_texto = corpus.raw()
todo_texto
```

```python
# Obtenção de todas as palavras do corpus e visualização da quantidade
palavras = corpus.words()
# Acessando pelo indice
palavras[1000]
```

```python
# quantidade
len(palavras)
```

```python
# Usando o nltk , obtemos as stopwords em inglês
stops = stopwords.words('english')
# stops = stopwords.words('portuguese')
stops
# tamanho de stops
len(stops)
```

```python
# Definição das cores que serão utilizadas na nuvem de palavras
mapa_cores = ListedColormap(['orange', 'green', 'magenta', 'red'])
# Criação da nuvem de palavras, com no máximo 100 palavras e utilizando as stopwords
nuvem = WordCloud(background_color='white',
                  colormap=mapa_cores,
                  stopwords=stops,
                  max_words=100)
# Criação e visualização da nuvem de palavras
nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
# Criação de nova lista de palavras, removendo stopwords
palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
# Remoção da pontuação, gerando uma lista sem stop words e sem pontuação
palavras_sem_pontuacao = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sem_pontuacao)
```

```python
# Cálculo da frequencia das palavras e visualização das mais comuns
frequencia = nltk.FreqDist(palavras_sem_pontuacao)
frequencia
```

```python
# mais comuns
mais_comuns = frequencia.most_common(100)
mais_comuns
```
