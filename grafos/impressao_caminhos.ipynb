```python
# Import de bibliotecas
from igraph import Graph, plot
```

```python
# criação de um grafo direcionado com pesos entre as arestas
grafo = Graph(edges = [(0,2), (0,1), (1,4), (1,5), (2,3), (6,7), (3,7), (4,7), (5,6)], directed = True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2, 1, 2, 1, 2, 1, 3, 1]
```

```python
# Visualização do grafo
plot(grafo, bbox = (300,300), edge_label = grafo.es['weight'])
```

```python
# Menor caminho entre os vértices A e H
caminho_vertice = grafo.get_shortest_paths(0,7, output = 'vpath')
caminho_vertice
```

```python
# Retorna as arestas que ligam os vértices A e H
caminho_aresta = grafo.get_shortest_paths(0, 7, output = 'epath')
caminho_aresta
```

```python
# Mostra o id das arestas que fazem parte do caminho
caminho_aresta_id = []
for n in caminho_aresta[0]:
    caminho_aresta_id.append(n)
caminho_aresta_id    
```

```python
# Mostra o nome dos certices que fazem parte do caminho
caminho_nome_vertices = []
for v in caminho_vertice[0]:
    print(grafo.vs[v]['label'])
    caminho_nome_vertices.append(grafo.vs[v]['label'])
caminho_nome_vertices    
```

```python
# Colorir so vértices do caminho
for v in grafo.vs:
    print(v)
    if v['label'] in caminho_nome_vertices:
        v['color'] = 'green'
    else:
        v['color'] = 'red'
plot(grafo, bbox = (300, 300), edge_label = grafo.es['weight'])                
```

```python
# Colorir as arestas do caminho
for e in grafo.es:
    if e.index in caminho_aresta_id:
        e['color'] = 'blue'
    else:
        e['color'] = 'grey'
plot(grafo, bbox = (300, 300), edge_label = grafo.es['weight'])
```
