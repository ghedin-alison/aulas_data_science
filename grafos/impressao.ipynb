```python
# Import das bibliotecas
from igraph import Graph, plot
```

```python
# Criação de grafo com pesos entre as relações
grafo = Graph(edges = [(0,1), (2,3), (0,2), (0,3),], directed = True)
grafo.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio']
grafo.vs['peso'] = [80, 83, 75, 68]
grafo.es['TipoAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo.es['weight'] = [1, 2, 1, 3]
```

```python
# Visualizar informações sobre os vértices
for v in grafo.vs:
    print(v)
```

```python
# Visualizar informações sobre as arestas
for e in grafo.es:
    print(e)
```

```python
# Definição de cores para os vértices
grafo.vs['cor'] = ['blue', 'red', 'yellow', 'green']
plot(grafo, bbox = (0, 0, 300, 300), vertex_color = grafo.vs['cor'])
```

```python
# Pesos para as arestas
plot(grafo, bbox = (300,300), edge_width = grafo.es['weight'], vertex_color = grafo.vs['cor'])
```

```python
# Pesos para os vertices
plot(grafo, bbox = (300, 300), vertex_size = grafo.vs['peso'], vertex_color = grafo.vs['cor'], edge_width = grafo.es['weight'])
```

```python
# Adicionando curvatura
plot(grafo, bbox = (300, 300), 
     vertex_size = grafo.vs['peso'], 
     edge_width = grafo.es['weight'],
     vertex_color = grafo.vs['cor'],
     edge_curved = 0.3)
```
