```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Girth com volume
plt.scatter(base.Girth, base.Volume)
```

```python
# height com volume
plt.scatter(base.Height, base.Volume, marker = '*')
```

```python
# Girth com height 
plt.scatter(base.Girth, base.Height)
```

```python
# histograma de volume
plt.hist(base.Volume)
plt.title('Arvores')
plt.xlabel('Volume')
plt.ylabel('Ocorrência')
```

```python
# imprimindo juntos
# Criação de figura, na qual os graficos serão posicionados
plt.figure(1)
plt.subplot(2, 2, 1) #(numero de linhas, numero de colunas, posição do grafico a seguir)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2, 2, 2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2, 2, 3)
plt.scatter(base.Height, base.Volume, marker = '*')
plt.subplot(2, 2, 4)
plt.hist(base.Volume)
```
